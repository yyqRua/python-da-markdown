```python
'''https://zhuanlan.zhihu.com/p/93423829'''
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
data = pd.read_csv('.\\test.csv', index_col=0, parse_dates=True)
```

# matplotlib入门: plt/ax/fig

```python
# fig创建画布；ax创建axes
fig, ax = plt.subplots(2,2,figsize = (16,9))
ax[0, 0].plot(data['BLUE_GREEN'])
ax[1, 0].plot(data['CRYPTOPHYTA'])
ax[0, 1].plot(data['DIATOM'])
ax[1, 1].plot(data['GREEN_ALGAE'])
ax[0,0].set_title('Title',fontsize=18)
```
